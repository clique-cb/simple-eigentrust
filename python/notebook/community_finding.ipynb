```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.fast_gnp_random_graph(100, 0.1)
```

```python
nx.draw(G)
```

```python
comms = nx.community.asyn_fluidc(G, 5)

community_labels = np.zeros(100, dtype=np.int16)
for k, c in enumerate(comms):
    for i in c:
        community_labels[i] = k
```

```python
nx.draw(G, node_color=[community_labels[i] for i in range(100)], cmap='tab20', node_size=100)
```

```python
k_clique_comms = nx.community.k_clique_communities(G, 3)
cms = list(k_clique_comms)
len(cms)
```

```python
plt.bar(range(len(cms)), [len(c) for c in cms])
```

```python
def membership_array(communities, n=100):
    membership = np.zeros((n, len(communities)), dtype=np.int16)
    for k, c in enumerate(communities):
        for i in c:
            membership[i, k] = 1
    return membership 
```

```python
comm_colors = np.random.rand(len(cms), 3)

# comm_membership_colors = None
#comm_colors
plt.imshow(np.ones((256, 256, 3)) * comm_colors[5])
```

```python
plt.imshow(membership_array(cms, 100), aspect='auto', cmap='Greys', interpolation='none')
```
